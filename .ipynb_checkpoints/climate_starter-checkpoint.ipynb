# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import inspect, create_engine, func
```

```python
engine = create_engine('sqlite:///Resources/hawaii.sqlite')
insp = inspect(engine)
```

```python
# reflect an existing database into a new model
db = automap_base()

# reflect the tables
db.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
db.classes.keys()
```

```python
# Save references to each table
Measurement = db.classes.measurement
Station = db.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
insp.get_table_names()
```

```python
columns = insp.get_columns('measurement')
for column in columns:
    print(column['name'], column['type'])
```

```python
measure_view = session.query(Measurement.id, Measurement.station, Measurement.date, Measurement.prcp, Measurement.tobs).all()
measure_view_df = pd.DataFrame(measure_view)
measure_view_df.head(10)
```

```python
columns = insp.get_columns('station')
for column in columns:
    print(column['name'], column['type'])
```

```python
station_view = session.query(Station.id, Station.station, Station.name, Station.latitude , Station.longitude, Station.elevation).all()
station_view_df = pd.DataFrame(station_view)
station_view_df.head(10)
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

last_date = session.query(Measurement.date).order_by(Measurement.date.desc())
last_date_df = pd.DataFrame(last_date)
last_date_df.head(5)
```

```python
first_date = session.query(Measurement.date).order_by(Measurement.date)
first_date_df = pd.DataFrame(first_date)
first_date_df.head(5)
```

```python
# Calculate the date 1 year ago from the last data point in the database
#Find all of the dates great than 2016-08-23
session.query(Measurement.date).\
    filter(Measurement.date >= '2016-08-23').\
    order_by(Measurement.date).count()#.all()for list
```

```python
query_date = dt.date(2017, 8, 23) - dt.timedelta(days=366)
print("Query Date: ", query_date)
```

```python
# Perform a query to retrieve the data and precipitation scores
```

```python
prcp_results = session.query(Measurement.date, Measurement.prcp).\
               filter(Measurement.date >= query_date).all()  
prcp_results               
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column # Sort the dataframe by date
prcp_results_df = pd.DataFrame(prcp_results).sort_values("date",ascending=True).dropna()
prcp_results_df 
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
prcp_results_df.plot('date', 'prcp', linewidth=8, figsize=(30,15), fontsize=20)
plt.xlabel("Date", fontsize=40)
plt.ylabel("Rain (inches)", fontsize=40)
plt.legend(["Precipitation"], loc="upper center", fontsize=40)
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig("precipitation year.png")
plt.show()
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?
busy_station = session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).filter(Measurement.station == 'USC00519281').all()
busy_station
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station #plot the results as a histogram
temp_obs = session.query(Measurement.tobs, Measurement.station).\
               filter(Measurement.date >= query_date).filter(Measurement.station == "USC00519281").all()
        
```

```python
temp_obs_df = pd.DataFrame(temp_obs)
temp_obs_df.set_index('station')
temp_obs_df
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```
